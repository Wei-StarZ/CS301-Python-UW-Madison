```python
import csv
import numpy as np
import matplotlib.pyplot as plt
```

```python
import math
x = np.arange(-3.5, 3.5, 0.01).tolist()
x
```

```python
noise = np.random.normal(0, 3, len(x))
plt.plot(noise)
plt.show()
y = np.zeros(len(x))
y
```

```python
it = range(0, len(x))
print(len(it))
print(len(x))
for i in it:
    y[i] = pow(x[i],3)- 4*math.log(abs(x[i]))*pow(x[i],2)/math.exp(1) + noise[i]
    
plt.plot(y)
plt.show()
```

```python
lines = np.zeros((len(x), 2))
for i in it:
    lines[i][0] = x[i]
    lines[i][1] = y[i]
    
print('lines =', lines)
print("x: ", x)
```

```python
with open('aperiodicCurve.csv', 'w',newline="") as writeFile:
    writer = csv.writer(writeFile)
    writer.writerows(lines)
    
writeFile.close()
```
